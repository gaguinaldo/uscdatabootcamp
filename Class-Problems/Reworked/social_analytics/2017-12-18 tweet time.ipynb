```python
tweet_time = []

for tweet in public_tweets:
    tweet_time.append(tweet['created_at'])
tweet_time 
```

```python
clean_date_time = []
for time_ in tweet_time:
    clean_date_time.append(datetime.strptime(time_, '%a %b %d %H:%M:%S %z %Y'))
```

```python
format_date_time = []
for time_ in clean_date_time:
    format_date_time.append(time_.strftime("%Y-%m-%d %H:%M:%S"))
```
